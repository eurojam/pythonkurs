**Einige String Operationen:**  
s.find(t)	 	Index erstes Vorkommen t in s (-1=nicht gefunden)  
s.rfind(t) 		index letztes Vorkommen t in s (-1=nicht gefunden)  
s.index(t) 		wie s.find(t) erzeugt ValueError wenn nicht gefunden  
s.rindex(t) 	wie s.rfind(t) erzeugt ValueError wenn nicht gefunden  
s.split(t) 		Splittet s mit t (default = Leerzeichen)  
s.lower() 		kleinschreibung  
s.upper() 		GROßSCHREIBUNG  
s.title() 		Erste Buchstabe Groß Geschrieben  
s.strip() 		Kopie von s ohne führende und anhängende Leerzeichen  
s.replace(t, u) Ersetzt t mit u im String s  
s.join(sequence)Verbindet die Elemente von sequence mit s  
len(s)			Anzahl Zeichen  
s+t				Zusammenfügen von s und t  
s*i				i Mal Wiederholen von s  
s[i:j]			Indexing von s, negative Indizes von hinten
