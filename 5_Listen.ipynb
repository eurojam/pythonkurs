# Datenstrukturen: Listen

Eine Liste ist eine geordnete Menge von Werten, wobei jeder Wert mit einem Index angesprochen wird. Die Werte, die die Liste bilden, werden ihre Elemente genannt. Listen haben Ähnlichkeiten mit Strings, die geordnete Mengen von Zeichen sind.

```python
# -*- coding: utf-8 -*-
a = [66.25, 333, 333, 1, 1234.5] # eine Liste
print(a.count(333), a.count(66.25), a.count('x')) #Anzahl Vorkommen eines Elements
a.insert(2, -1) #Einfügen vor dem Index
a.append(333) #Am Ende anfügen
print (a)
print (a.index(333)) #Index erstes Vorkommen
a.remove(333) #Enfernt erstes Vorkommen
print (a)
a.reverse() #Umkehr der Liste (in place!)
print (a)
a.sort() #Sortieren (in place!)
print (a)
```

**Liste als Stack verwenden: last-in, first-out**

```python
stack = [3, 4, 5]
stack.append(6) #Element auf den Stack legen
stack.append(7)
print (stack)
stack.pop() #Element vom Stack nehmen
print (stack)
```

**Liste als Queue verwenden: first-in, first-out**

```python
#Verwendung des Moduls deque
from collections import deque
queue = deque(["Erik", "Hans", "Michael"])
queue.append("Dirk")
queue.append("Gerhard")
print (queue)
queue.popleft() #Erik geht
queue.popleft() #Hans geht
print (queue)
```

**List Comprehensions (Listen-Abstraktion)**
ist eine elegante Methode Mengen in Python zu erzeugen

```python
squares = []
for x in range(10):
    squares.append(x**2)
print (squares)

squares2 = [x**2 for x in range(10)] #List Comprehensions
print (squares2)
```

```python
squares = []
for x in range(10):
    squares.append(x**2)
    
# etwas kürzer geht das so
squares2 = [x**2 for x in range(10)] #List Comprehensions

#List Comprehensions mit komlexeren Ausdrücken
from math import pi
pi_round = [str(round(pi, i)) for i in range(1, 6)]
#Die Ergebnisse sehen wie folgt aus
print (squares, '\n', squares2, '\n', pi_round)

# Löschen geht neben pop/remove auch mit del
a = [-1, 1, 66.25, 333, 333, 1234.5]
del a[0]
print (a)
del a[2:4]
print (a)
[1, 66.25, 1234.5]
del a[:]
print (a)
```

**Allgemeine Syntax von List Comprehensions**  
Beispiel: new_range  = [i * i for i in range(5) if i % 2 == 0]  

**Allg. Syntax:** *result*  = [*transform*    *iteration*         *filter*]

```python
# alle Zahlenpaare, deren Summe 7 ist:
[(x,y) for x in range(1,7) for y in range (1,7) if x+y==7]
```

```python
# alle Zahlen, die durch 3 teilbar sind
[x for x in range(1,40) if  x % 3 == 0]
```

```python
# Primzahlen bestimmen
noprimes = [j for i in range(2,8) for j in range(i*2, 100, i) ]
primes = [x for x in range(2,100) if x not in noprimes]
primes
```

```python
#ein Kreuzprodukt
# -*- coding: utf-8 -*-
farben = ["rot", "gelb", "blau"]
dinge = ["Auto", "Boot", "Fahrrad", "Haus"]
farbige_dinge = [(x,y) for x in farben for y in dinge if x=='rot']
print (farbige_dinge)
```
